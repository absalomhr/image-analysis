#### What is image scaling
Image scaling refers to the resizing of a digital image. When scaling a raster graphics image, a new image with a higher or lower number of pixels must be generated. This usually results in a visible quality loss.
From the standpoint of digital signal processing, the scaling of raster graphics is a two-dimensional example of sample-rate conversion, the conversion of a discrete signal from a sampling rate (in this case the local sampling rate) to another (the desired one, whether it's larger or shorter).
#### Linear resampling algorithm (interpolation)
<img src="./img/interpolation.png" alt="escom logo" align="center" width="375"/>
In mathematics, linear interpolation is a method of curve fitting using linear polynomials to construct new data points within the range of a discrete set of known data points. 
<img src="./img/interpolation2.png" alt="escom logo" align="center" width="250"/>
If the two known points are given by the coordinates $(x_{0}, y_{0})$ and $(x_{1}, y_{1})$, .<strong>the linear interpolant</strong> is the straight line between these points. For a value $x$ in the interval $(x_{0},x_{1})$, the value $y$ along the straight line is given from the equation of slopes:
$${\frac {y-y_{0}}{x-x_{0}}}={\frac {y_{1}-y_{0}}{x_{1}-x_{0}}}$$
Solving this equation for y, which is the unknown value at x, gives:
$$y=y_{0}+(x-x_{0}){\frac {y_{1}-y_{0}}{x_{1}-x_{0}}}={\frac {y_{0}(x_{1}-x)+y_{1}(x-x_{0})}{x_{1}-x_{0}}}$$

#### Algorithm
Here we take as an example of scaling an old image of $h \cdot w$ to a new image of $h \cdot w_n$, where $w_n \neq w$

- In order to do the linear interpolation we need to work in the same scale thats why we need to divide the $\frac{w}{w_n}$ so we can know the value of $x$ in respect to the old image for each pixel in the new image
- Now that we know the value of $x$ we can easily calculate the respective value $y$ for each point because we have the $y$ values of the old images

```python
def scale_img(n_height, n_width, c_height, c_width, channels, n_img, o_img):
    # Width scaling first
    for row in range(c_height):
        for pixel in range(n_width):
            # Point in the x axis according to the new scale
            x = (c_width - 1) * pixel / n_width
            x0 = int(x)
            x1 = x0 + 1
            for ch in range(channels):
                y0 = int(o_img[row][x0][ch])
                y1 = int(o_img[row][x1][ch])
                # Equation used for finding y
                y = y0 + (x - x0) * (y1 - y0) / (x1 - x0)
                # Point in the y axis according to the line that goes from (x0, y0) to (x1, y1)
                n_img[row][pixel][ch] = int(round(y))
    temp_img = np.copy(n_img)
    # Height scaling
    for col in range(n_width):
        for pixel in range(n_height):
            x = (c_height - 1) * pixel / n_height
            x0 = int(x)
            x1 = x0 + 1
            for ch in range(channels):
                y0 = int(temp_img[x0][col][ch])
                y1 = int(temp_img[x1][col][ch])
                # Equation used for finding y
                y = y0 + (x - x0) * (y1 - y0) / (x1 - x0)
                # Point in the y axis according to the line that goes from (x0, y0) to (x1, y1)
                n_img[pixel][col][ch] = int(round(y))
            
```

#### Choosing the desired size for the new image

```python
# The current size
c_height = img_arr.shape[0]
c_width = img_arr.shape[1]
# The desired size of the image
n_height = c_height * 2
n_width = c_width * 2
# New image
n_img = np.zeros((n_height, n_width, img_arr.shape[2]), dtype=int)
%time scale_img(n_height, n_width, c_height, c_width, n_img.shape[2], n_img, img_arr)
plt.axis("off")
plt.title("Current image size: " + str(n_img.shape[:2]))
plt.imshow(n_img)
```

#### Now we call the funcion again with the same image and the same new size to compare execution time

```python
# The current size
c_height = img_arr.shape[0]
c_width = img_arr.shape[1]
# The desired size of the image
n_height = c_height * 2
n_width = c_width * 2
# New image
n_img = np.zeros((n_height, n_width, img_arr.shape[2]), dtype=int)
%time scale_img_c(n_height, n_width, c_height, c_width, n_img.shape[2], n_img, img_arr)
plt.axis("off")
plt.title("Current image size: " + str(n_img.shape[:2]))
plt.imshow(n_img)
```
